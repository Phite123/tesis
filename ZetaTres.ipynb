```python
# ============================================================
# 1. Generic fiber over Q(sqrt(-3))(k)
# ============================================================

Qw.<w> = QuadraticField(-3)

R.<k> = PolynomialRing(Qw)
K = FractionField(R)

D = K(k^2 - 1)

E = EllipticCurve(K, [0, 0, 0, 0, D^2])

print("Generic curve:")
print(E)

print("\nInvariants:")
print("Discriminant:", factor(E.discriminant()))
print("c4:", E.c4())
print("c6:", factor(E.c6()))
print("j:", E.j_invariant())

xP = D
yP = K(k)*D
P = E([xP, yP])
print("P =", P)

# ============================================================
# 2. Specialization k = k0 in Q(sqrt(-3))
# ============================================================

OK = Qw.ring_of_integers()

def specialize_E(k0):
    """
    Specialize the elliptic surface

        y^2 = x^3 + (k^2 - 1)^2

    at k = k0 in Q(sqrt(-3)).
    """
    k0 = Qw(k0)
    D0 = k0^2 - 1

    if D0 == 0:
        raise ValueError("Singular fiber: k0^2 - 1 = 0.")

    E0 = EllipticCurve(Qw, [0, 0, 0, 0, D0^2])
    P0 = E0([D0, k0*D0])

    return E0, P0

# ============================================================
# 3. Local arithmetic over Q(sqrt(-3))
# ============================================================

def bad_prime_candidates(E):
    """
    Candidate bad primes from the support of the discriminant ideal.
    Works over number fields.
    """
    OK = E.base_field().ring_of_integers()
    I = OK.fractional_ideal(E.discriminant())
    return [Pideal for Pideal, e in I.factor()]


def bad_primes(E):
    """
    Try to filter the support of the discriminant to actual bad primes.
    If Sage cannot decide with has_good_reduction, keep the prime.
    """
    ans = []

    for Pideal in bad_prime_candidates(E):
        try:
            if not E.has_good_reduction(Pideal):
                ans.append(Pideal)
        except Exception:
            ans.append(Pideal)

    return ans


def local_report(E):
    OK = E.base_field().ring_of_integers()

    print("\nCurve:")
    print(E)

    print("\nDiscriminant:")
    print(E.discriminant())

    print("\nDiscriminant ideal factorization:")
    print(OK.fractional_ideal(E.discriminant()).factor())

    bad = bad_primes(E)

    print("\nCandidate bad primes:")
    print(bad)

    for Pideal in bad:
        print("\n" + "="*70)
        print("Prime ideal:")
        print(Pideal)

        try:
            print("\nLocal data:")
            print(E.local_data(Pideal))
        except Exception as e:
            print("Could not compute local_data:", e)

        try:
            print("Kodaira symbol:", E.kodaira_symbol(Pideal))
        except Exception as e:
            print("Could not compute Kodaira symbol:", e)

        try:
            print("Tamagawa number:", E.tamagawa_number(Pideal))
        except Exception as e:
            print("Could not compute Tamagawa number:", e)


def test_specializations(values):
    for k0 in values:
        print("\n" + "#"*80)
        print("k =", k0)

        try:
            E0, P0 = specialize_E(k0)
        except ValueError as e:
            print(e)
            continue

        print("P0 =", P0)
        print("P0 on E0?", P0 in E0)

        try:
            print("Order of P0:", P0.order())
        except Exception as e:
            print("Could not compute order:", e)

        local_report(E0)


values = [
    QQ(1)/2 + w,
    QQ(3)/2 - QQ(1)/3*w,
    35 - 21*w,
    332 - 532*w,
    65 - 2*w,
    423 - 843*w
]

test_specializations(values)
```

```python
Qi.<i> = QuadraticField(-1)               # Q(i)          OJO con poner algo como for i in range(), sino poner ii a la variable
Ri.<r,s> = PolynomialRing(Qi)             # Q(i)[r,s]
Ki = FractionField(Ri)                    # Q(i)(r,s)
T.<T> = PolynomialRing(Ki)                # Ki[T] para agregar raiz de T^2 - (r^2 - s^2) osea asumir que r^2 - s^2 es un cuadrado en el cuerpo
Li.<u> = Ki.extension(T^2 - (r^2 - s^2))  # u^2 = r^2 - s^2.
r = Li(r); s = Li(s); i = Li(i)           # Reasignamos r, s, i como elementos de Li

Ers = EllipticCurve(Li, [0, r^2 + s^2, 0, r^2*s^2, 0])

P = Ers([r*s, r*s*(r+s)])
Q = Ers([r*(u-r), i*r*u*(u-r)])
O = Ers([0,1,0])

def limpio(expr):
    return (
        str(expr)
        .replace("*", "")
        .replace("(-i)", "-i")
    )
def punto(R):
    if R == O:
        return r"\OO"
    return f"({limpio(R[0])},{limpio(R[1])})"

for j in range(4):
    print(" & ".join(punto(k*P + j*Q) for k in range(4)) + r" \\ \hline")

```
